```python
#Importing the required modules
import pandas as pd

#Importing the required data
hep_data = pd.read_csv('HepatitisCdata.csv')
hep_data.head()

#Creating index1 column in hep_data to have a commom column
hep_data['index1'] = hep_data.index
print(hep_data.head())

hep_data.shape
```

```python
hrt_data = pd.read_csv('heart.csv')
hrt_data.head()

#Creating index1 column in hrt_data
hrt_data['index1'] = hrt_data.index
print(hrt_data.head())

hrt_data.shape
```

```python
#Merging hep_data and hrt_data with the default merge (inner)
hep_hrt1 = pd.merge(hep_data, hrt_data, on='index1')
hep_hrt1

```

```python
#Merged dataset
hep_hrt1.shape
```

```python
#Merging hep_data and hrt_data with the default merge (outer)
hep_hrt2 = pd.merge(hep_data, hrt_data, on='index1', how='outer', suffixes= ('_hep', '_hrt'))
hep_hrt2

```

```python
#Merged dataset
hep_hrt2.shape
```

```python
#Merging hep_data and hrt_data with the default merge (left)
hep_hrt3 = pd.merge(hep_data, hrt_data, on='index1', how='left', suffixes= ('_hep', '_hrt'))
pd.set_option('display.max_columns', None)
hep_hrt3

```

```python
#Merged dataset
hep_hrt3.shape
```

```python
hep_hrt4 = pd.merge(hep_data, hrt_data, on='index1', how='right', suffixes= ('_hep', '_hrt'))
pd.set_option('display.max_columns', None)
hep_hrt4
```

```python
hep_hrt4.shape
```

```python
hep_hrt_cc = pd.concat([hep_data, hrt_data], axis=1)
hep_hrt_cc
```
